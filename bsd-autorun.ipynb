```python
A = 0
B = 7
N = 100000
```

```python
def counting_points(p): 
    try:
        E = EllipticCurve(GF(p),[A,B]); E
        return E.cardinality() 
    except ArithmeticError:
        #print('ArithmeticError')
        s =0
        for a in range(p):
            for b in range(p):
                if (b*b - (a*a*a+A*a +B) ) % p == 0:
                    s += 1 
        return s+1
    
p= 2
data = [] 
product = 1

P = Primes() 
while p < N:
    n = counting_points(p)
#print(p, n)
# #E_p(F_p) 計算 n 代入
    product *= n/p
    data.append( (log(p), product) ) #print("")
    p = P.next(p)
    
print('finished')
pl = point(data) 
pl.show(scale='loglog')
```
